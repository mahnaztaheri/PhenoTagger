```python
import pandas as pd

def parse_tsv_to_pubtator(tsv_file, output_file):

    with open(tsv_file, 'r') as file:
        lines = file.readlines()


    pmid = None
    abstract = None
    annotations = []


    with open(output_file, 'w') as out_f:
        for line in lines:
            line = line.strip()

            if line.isdigit():
                if pmid and abstract:
                    out_f.write(f"{pmid}|t|{abstract}\n\n")
                pmid = line
                abstract = None
                annotations = []

            elif abstract is None:
                abstract = line

        if pmid and abstract:
            out_f.write(f"{pmid}|t|{abstract}\n\n")


```
